```python
import pydicom
import os

for root, dirs, files in os.walk("siim"):

    for file in files:

        if file.endswith(".dcm"):

            ds = pydicom.dcmread(
                os.path.join(root,file),
                stop_before_pixels=True
            )

            print("Modality:",
                  ds.get("Modality"))

            print("Body Part:",
                  ds.get("BodyPartExamined"))

            break

    break
```

```python
count = 0

for root, dirs, files in os.walk("siim"):
    for file in files:
        if file.endswith(".dcm"):
            count += 1

print("DICOM files found:", count)
```

```python
import os
import pydicom
from collections import Counter

modalities = Counter()

for root, dirs, files in os.walk("siim"):

    for file in files:

        if file.endswith(".dcm"):

            try:
                ds = pydicom.dcmread(
                    os.path.join(root,file),
                    stop_before_pixels=True
                )

                modality = ds.get(
                    "Modality",
                    "Unknown"
                )

                modalities[modality] += 1

            except:
                pass

print(modalities)
```
